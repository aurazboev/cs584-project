```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

data = {'Day':[11,14,5,14,10],'Month':[1,1,8,8,11],'Year':[2022,2022,2022,2022,2022],'Hour':[15,15,21,14,3],'Location Description':['RESIDENCE','RESIDENCE','APARTMENT','RESIDENCE','STREET'],'Community area':[46.0,68.0,69.0,11.0,68.0],'Primary Type':['SEX OFFENSE','OTHER OFFENSE','SEX OFFENSE','SEX OFFENSE','WEAPONS VIOLATION']}
data_2 = {'Day':[14,10],'Month':[8,11],'Year':[2022,2022],'Hour':[14,3],'Location Description':['RESIDENCE','STREET'],'Community area':[11.0,68.0],'Primary Type':['SEX OFFENSE','WEAPONS VIOLATION']}
df = pd.DataFrame(data)
df_2 = pd.DataFrame(data_2)

le = LabelEncoder()
df['Primary Type'] = le.fit_transform(df['Primary Type'])
df_2['Primary Type'] = le.transform(df_2['Primary Type'])

onehot_encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')  
df_location_encoded = pd.DataFrame(onehot_encoder.fit_transform(df[['Location Description']]), columns=onehot_encoder.get_feature_names_out(['Location Description']))
df = pd.concat([df, df_location_encoded], axis=1)

df_2_location_encoded = pd.DataFrame(onehot_encoder.transform(df_2[['Location Description']]), columns=onehot_encoder.get_feature_names_out(['Location Description']))
df_2 = pd.concat([df_2, df_2_location_encoded], axis=1)

features = ['Day', 'Month', 'Year', 'Hour', 'Community area'] + list(df_location_encoded.columns)
target = 'Primary Type'

models = {
    'Multinomial Logistic Regression': LogisticRegression(multi_class='multinomial', max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Na√Øve Bayes': MultinomialNB(),
    'K-nearest Neighbors': KNeighborsClassifier(),
    'Support Vector Machine': SVC(),
    'Neural Network': MLPClassifier(max_iter=1000)
}

for name, model in models.items():
    model.fit(df[features], df[target])
    predictions_df_2 = model.predict(df_2[features])
    predicted_labels = le.inverse_transform(predictions_df_2)   
    print("Model: ",name)
    print("Predictions for df_2:")
    print(predicted_labels)
    print("Accuracy:", accuracy_score(df_2['Primary Type'], predictions_df_2))
    print("Classification Report:\n\n", classification_report(df_2['Primary Type'], predictions_df_2))
    print("\n" + "-"*55 + "\n")
```
